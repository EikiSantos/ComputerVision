### 2) PETs Computational Vision Group:
O PETS CVG (Performance Evaluation of Tracking and Surveillance - Computational Vision Group) ocorre anualmente, junta um conjunto de pesquisadores que trabalham na área de vigilancia visual e trabalham em cima de um unico Dataset, com o intuito de possibilitar a avaliação sistemática do desempenho das técnicas desenvolvidas, visto que é difícil fazer comparações entre algoritmos que foram testados em diferentes conjuntos de dados com condições variáveis.


### 2000: 

#### <font color="red"> a) Dataset: 
Pessoas ao ar livre e rastreamento de veículos registrados por uma única camêra. Cenas de área ampla. Uma sequência de imagens de um ambiente de estacionamento contendo pessoas em movimento e veículos.
    
#### <font color="red"> b) Resumo:
Obteve-se alguns métodos de <font color="blue"> rastreamento de veículos</font>, como: 

1. Rastreamento sem detecção de recursos.
2. Rastreamento de Veículos usando Estimativa de Fluxo Óptico Adaptativo. 
3. Compressão de imagens para refinamento da trajetória e Modelos de formas de memória curta para rastreamento de objetos preditivos em plano terrestre.
    
Também obteve-se alguns métodos de <font color="blue"> rastreamento de pessoas e veículos </font>, como: 

1. Rastreamento sem modelo de carros e pessoas com base em regiões de cores.
2. detecção e rastreamento de movimento usando Belief Indicators para aplicações de vigilância por vídeo.

### 2001: 
#### <font color="red"> a) Dataset:
Pessoas ao ar livre e veículos registrados por 2 visualizações sincronizadas, incluindo camera omidirecional e em movimento.
    
#### <font color="red"> b) Resumo:
O PETS2001 deu continuidade ao tema do PETS2000, porém com um dataset diferente e mais complexo.
 
Obteve-se métodos de <font color="blue">rastreamento baseado em aparência</font>, como: 

1. Rastreamento de alvos em interação usando aproximações gaussianas.
2. Rastreamento de aglomerados não gaussianos nas sequências de imagens do dataset.
3. Rastreamento de pessoas em aplicações de vigilância.
4. Modelos de aparência para tratamento de oclusão.

Obteve-se métodos de <font color="blue">rastreamento de pessoas e veículos</font>, como:

1. Deteclçao e Rastreamento de objetos em um mundo aberto e dinâmico.
2. Rastreamento em tempo real de pedestres e veículos.
3. Rastreamento e Classificação de objetos móveis em video.
4. Rastreamento de pessoas e veículos em sequências de imagens externas usand transformação temporal de velocidade espacial.

Obteve-se métodos de <font color="blue">rastreamento de múltiplas visualizações</font>, como:

1. Rastreamento de Imagem de Múltiplas Câmeras.
2. Rastreamento baseado em cores de vários objetos usando várias câmeras em cenas externas e complexas que variam no tempo.
3. Rastreamento em câmeras não calibradas com campo de visão sobreposto.

### 2004: 
#### <font color="red"> a) Dataset:
Videoclipes representando os diferentes cenários de interesse: pessoas caminhando sozinhas, reuniões com outras pessoas, vitrines, brigas e desmaios, ou deixar uma mala em um local público. Todos os videoclipes foram filmados para o projeto CAVIAR, com uma lente grande angular no sagão de entrada do INRIA Labs, em Grenoble, França.
    
#### <font color="red"> b) Resumo:

Foram abordados os seguintes tópicos sobre <font color ="blue">Regulação Automática</font>:

1. Rastreamento visual robusto do controle dinâmico de processamento.
2. Rastreamento automático e rotulagem de atividade humanas em uma sequêmcia de vídeos.
3. Associação de dados baseada em aparência e trajetória conjunta para rastreamento de objetos múltilos

Foram abordados os seguintes tópicos sobre <font color="blue">Semântica</font>:

1. Isto é interessante? Comparando julgamentos humanos e de máquina no conjunto de dados PETS.
2. Treinamento guiado por ontologia de redes bayesianas para análise de alto nível em vigilância visual.
3. Grupos de rastreamento on-line de pedestres com redes bayesianas.

Foram abordados os seguintes tópicos sobre <font color="blue">Métricas</font>:

1. Novas métricas de avaliação de desempenho para algoritmos de detecção de objetos.
2. Métrica de distância de trajetória usando representação baseada em modelo Markov oculto.

### 2006: 

#### <font color="red"> a) Dataset:
Sequências de multissensores que contêm cenários de left luggage com crescente complexidade da cena.

#### <font color="red"> b) Resumo:
1. Definição de "left luggage": a bagagem de mão no contexto do PETS 2006 é definida como itens de bagagem que foram abandonados pelo proprietário e cada pessoa possui no máximo um item de bagagem. Itens classificados como bagagem são: Pasta, mala de viagem, mochila de 25 litros, mochila de 70 litros, portador de equipamento de esqui.
    
2. O que constitui uma bagagem esquecida?
Três regras são usadas para determinar se a bagagem é da pessoa ou não:

Regra Contextual: A pessoa entra em cena com a bagagem até o ponto em que a bagagem manteve contato fisico com a pessoa.

Regra Espacial: Nesse momento a bagage é atendida somente pelo proprietário quando eles estão a uma distância de um metro da bagagem
    <img src="luggage1.png">
    
Regra Espaço-Temporal: Um item é desacompanhado quando o proprietário está a uma certa distância em metros da bagagem. E permanece n segundos nessa configuração de evento.
    <img src="lugagge2.png">    
    
3. Foram Abordados os seguintes tópicos:

- Avaliação de segmentação e rastreamento de movimento.
- Detecção de left luggage usando homografias e heurísticas simples.
- Detecção e rastreamento automáticos de left luggage usando várias câmeras.
- Detecção de várias vistas e rastreamento de viajantes e bagagem em ambientes de transporte de massa.
- Detectando itens de bagagem abandonados em um espaço público.
- Detecção de Bagagem Esquerda usando Inferência Bayesiana.
- Avaliação de um sistema IVS para detecção de objetos abandonados em conjuntos de dados PETS2006.
- Detecção de objetos abandonados em locais lotados.
- Avaliação geral dos resultados do PETS2006.

### 2009: 
#### <font color="red"> a) Dataset: 
Análise de imagem de multidões, observada por uma camera unica ou uma rede de cameras.
    
#### <font color="red"> b) Resumo:  
A vigilância visual é uma importante área da visão computacional. O desafio científico na analise de imagens de multidões é analisar imagens de multidões e criar e implementar sistemas automáticos para obter informações detalhadas sobre os movimentos e o comportamento coletivo de indivíduos e gurpos em uma cena lotada observada por uma unica camera ou por uma rede de cameras. A análise das imagens de multidões incluem a estimativa de densidade e contagem, rastreamento de indivíduos dentro de uma multidão e detecção de fluxos separados e eventos específicos de multidões.
